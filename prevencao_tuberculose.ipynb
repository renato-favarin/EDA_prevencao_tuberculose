# 1. Introdução

A tuberculose é uma doença infectocontagiosa causada pelo *Mycobacterium tuberculosis* ou pelo Bacilo de Koch. Ela ataca mais comumente os pulmões, mas pode também causar infecções nos ossos, rins e meninges (as membranas que envolvem o cérebro).

A transmissão da tuberculose ocorre pelo meio aéreo, pessoa a pessoa, portanto aglomerações são a principal fonte de transmissão.

Algumas das consequências da doença são tosse seca, emagrecimento, falta de apetite e eliminação de sangue pela tosse, além de poder levar a óbito.

No panorama mundial, de acordo com a Organização Mundial da Saúde (OMS), ainda hoje a tuberculose está entre as dez maiores causas de morte no mundo, atingindo 10 milhões de casos por ano e mais de 1,5 milhão de óbitos.

O melhor combate à tuberculose é pela prevenção. A vacina BCG, administrada logo após o nascimento, nos protege de formas mais graves da doença; a meta federal para esta vacina é aplicar em mais de 90% da população-alvo.

Uma vez adquirida a doença, o tratamento é a base de antibióticos e outros medicamentos e é bastante demorado, levando cerca de 6 meses até a completa recuperação.

Antes da vacinação e dos antibióticos, até a década de 1940, o tratamento era basicamente repouso e boa alimentação nos chamados sanatórios.

Um desses sanatórios foi idealizado por Vicentina de Queiroz Aranha, esposa do senador Olavo Egídio, que atuava no combate à doença.  Mesmo após sua morte, em 1916, sua campanha prosseguiu e culminou na construção do sanatório Vicentina Aranha em São José dos Campos (SP). As obras iniciaram em 1918 e ele foi inaugurado em 1924, entrando em funcionamento mesmo antes de estar concluído, devido ao crescente número de casos. Os pacientes ficavam anos morando no sanatório, longe de seus familiares, até adquirir sua completa cura.

A escolha da cidade se deu devido ao clima, considerado ideal para o tratamento da doença.

Com o desenvolvimento de antibióticos na década de 1950, o tratamento tornou-se mais simples e não mais necessitava o isolamento em sanatórios. O paciente podia ser tratado em sua própria casa. Portanto, as estruturas dos complexos sanatoriais passaram a ser desnecessárias e pouco a pouco foram loteadas, dando origem à edifícios comerciais e residênciais. Mas foi apenas no início da década de 1980 que o sanatório encerrou suas atividades após o último paciente receber alta.

Atualmente é um agradável parque, cuja memória é preservada e respeitada.

## 1.1 A vacina BCG


**O que previne:**

Tuberculose – principalmente as formas graves, como meningite tuberculosa e tuberculose miliar (espalhada pelo corpo).

**Indicação:**

A vacina é indicada de rotina a partir do nascimento até antes de a criança completar 5 anos de idade.

Pessoas de qualquer idade que convivem com portadores de hanseníase (lepra).

**Resultados da prevenção no Brasil e no mundo:**

A vacina BCG não oferece eficácia de 100% na prevenção da tuberculose pulmonar, mas sua aplicação em massa permite a prevenção de formas graves da doença, como a meningite tuberculosa e a tuberculose miliar (forma disseminada).

No Brasil, embora a incidência de tuberculose pulmonar venha aumentando, quase não são mais registradas as formas graves da doença. Outro exemplo da importância da vacinação foi o aumento do número de casos de tuberculose em crianças, registrado quando a Suécia suspendeu a vacinação de rotina.

A Organização Mundial da Saúde (OMS) estima que, nos países onde a tuberculose é frequente e a vacina integra o programa de vacinação infantil, previna-se mais de 40 mil casos anuais de meningite tuberculosa. Impacto como este depende de alta cobertura vacinal, razão pela qual é tão importante que toda criança receba a vacina BCG.

Fonte: https://familia.sbim.org.br/vacinas/vacinas-disponiveis/vacina-bcg

## 1.2 Cobertura vacinal


A fórmula de cálculo da cobertura é o número de doses aplicadas da dose indicada (1ª, 2ª, 3ª dose ou dose única, conforme a vacina) dividida
pela população alvo, multiplicado por 100. Exemplo: para a Tetravalente (DTP/Hib), considera-se o número de terceiras doses aplicadas na faixa
etária de menores de 1 ano. Para a vacina oral de rotavírus humano, pode-se avaliar cobertura de 1ª e 2ª doses. <br>
O publico alvo da vacina BCG são bebês com menos de 1 ano. <br>
Fonte: http://tabnet.datasus.gov.br/cgi/pni/Imun_cobertura_desde_1994.pdf

<br>

As coberturas acima de 100% acontecem quando são aplicadas doses extras ou acima do previsto. <br>
Fonte: https://www.spdm.org.br/saude/noticias/item/3191-mundo-vive-maior-surto-global-de-sarampo-desde-2006-cobertura-vacinal-e-uma-das-mais-baixas-em-anos

```python
casos_confirmados = casos_confirmados.drop(columns = ["IG", "Total"])
casos_confirmados.info()
```

# Cálculo de casos de tuberculose por 100 mil habitantes e UF

Com os dataframes da população e dos casos de tuberculose preparados, é hora de calcular os casos por 100 mil habitantes.

```python
# dividir o df da população por 100k habitantes (como os dados de 2020 ainda não estão completos esta linha será descartada)

populacao_100k_hab = populacao.T[:-1]/100000
populacao_100k_hab.tail()
```

```python
# a divisão, nesse momento, do df "casos_confirmados" pelo df "populacao_100k_hab" resulta em erro.
# repare que um índice é formado por strings e o outro é formado por inteiros
# essa é a causa do erro

display(populacao_100k_hab.index)
print('\n')
display(casos_confirmados.index)
```

```python
# deixar os índices iguais - afinal eles estão na mesma ordem - e assim poder calcular a quantidade de casos de tuberculose por 100k habitantes
populacao_100k_hab.index = casos_confirmados.index

# divisão do dataframe de casos confirmados pelo dataframe da população de 100k habitantes
casos_por_100k_hab = casos_confirmados.div(populacao_100k_hab, axis='index')
casos_por_100k_hab.head()
```

```python
# abrir o dataframe para poder plotar
aberto_casos_por_100k_hab = casos_por_100k_hab.reset_index().melt(id_vars=["Ano Diagnóstico"])

# renomear as colunas
aberto_casos_por_100k_hab.columns = ["Ano", "UF", "Valor"]

# dicionário para poder incluir a região no dataframe
dic_regiao = {
    'AC':'Norte',
    'AL':'Nordeste',
    'AP':'Norte',
    'AM':'Norte',
    'BA':'Nordeste',
    'CE':'Nordeste',
    'DF':'Centro-Oeste',
    'ES':'Sudeste',
    'GO':'Centro-Oeste',
    'MA':'Nordeste',
    'MT':'Centro-Oeste',
    'MS':'Centro-Oeste',
    'MG':'Sudeste',
    'PA':'Norte',
    'PB':'Nordeste',
    'PR':'Sul',
    'PE':'Nordeste',
    'PI':'Nordeste',
    'RJ':'Sudeste',
    'RN':'Nordeste',
    'RS':'Sul',
    'RO':'Norte',
    'RR':'Norte',
    'SC':'Sul',
    'SP':'Sudeste',
    'SE':'Nordeste',
    'TO':'Norte',
}

# mapeamento do dataframe
aberto_casos_por_100k_hab["regiao"] = aberto_casos_por_100k_hab["UF"].map(dic_regiao)
aberto_casos_por_100k_hab
```

```python
# listas para iterar na função "casos_regiao"

regiao = ["Sudeste", "Sul", "Nordeste", "Centro-Oeste", "Norte"]
ufs = {"Sudeste": ['RJ','SP','ES', 'MG'], "Sul": ['RS','SC','PR'], "Nordeste":['PE','AL','PI','CE','BA','SE','MA','RN','PB'], "Centro-Oeste": ['MS','MT', 'GO','DF'], "Norte": ['AM','AC','PA','RR','RO','AP','TO'] }
```

```python
# lista dos gráficos para a última análise
graficos = []

# função para plotar cada um dos gráficos de casos/100k habitantes
def casos_regiao(regiao):
  aberto_casos_uf = aberto_casos_por_100k_hab.groupby(["Ano", "UF","regiao"]).sum("Valor")
  aberto_casos_uf = aberto_casos_uf.reset_index()
  sns.set_theme(style = "whitegrid", font='serif', color_codes=True, font_scale = 1.6)

  # plotar um gráfico de regressão polinomial de ordem 2
  ax = sns.lmplot(data = aberto_casos_uf[aberto_casos_uf["regiao"]==regiao], x="Ano", y = "Valor", hue="UF", legend=False, height=9,  order=2, scatter_kws={"s": 50}, line_kws={"lw":5})
  plt.grid(linestyle = '--')

  # alterar a frequência dos valores do eixo y
  plt.yticks(np.arange(0, 150, step=25))
  plt.xticks(rotation=0)

  plt.xlabel("")
  plt.ylabel("")
  plt.title("Casos de tuberculose por 100 mil habitantes no \n {} (2001 - 2019)".format(regiao))

  ax.add_legend(label_order = ufs[regiao], fontsize=17, title="", loc='upper center', ncol=4, borderaxespad=3.5, frameon = True)

  image_name = f'casos_regiao_{regiao}.png'
  graficos.append(image_name)
  plt.savefig(image_name, bbox_inches='tight', pad_inches=0.2)

  plt.show(ax)
  print("\n \n")
```

```python
# chamada da função para plotar os gráficos
for cada_regiao in regiao:
  casos_regiao(cada_regiao)
```

# Cálculo da cobertura vacinal por UF

Com os dataframes do número de nascidos vivos e do total de doses aplicadas de BCG, ambos preparados, é hora de calcular a cobertura vacinal por UF.

```python
# relembrando a fórmula da cobertura vacinal da BCG: 
# CV = [(número de doses aplicadas da dose indicada) / (população alvo)] * 100
# a dose indicada da BCG é única e a população-alvo são os recém-nascidos.

cobertura_vacinal = round((doses/nascidos_vivos * 100), 2)
cobertura_vacinal
```

Verifica-se que na maioria dos anos a cobertura ficou acima dos 100%. Isso acontece quando são aplicadas doses extras, por exemplo para população de outras idades, que não fazem parte da população-alvo inicial.

```python
# abrir o dataframe para poder plotar
aberto_cobertura_vacinal = cobertura_vacinal.reset_index().melt(id_vars=["Ano"])

# renomear as colunas
aberto_cobertura_vacinal.columns = ["Ano", "UF", "Valor"]

# mapeamento do dataframe (utilizando o mesmo dicionário criado anteriormente)
aberto_cobertura_vacinal["regiao"] = aberto_cobertura_vacinal["UF"].map(dic_regiao)
aberto_cobertura_vacinal
```

```python
# lista para iterar na função "casos_regiao"
ufs = {"Sudeste": ['RJ','SP','ES','MG'], "Sul": ['RS', 'SC','PR'], "Nordeste":['PE','AL','PI','CE','BA','SE','MA','RN','PB'], "Centro-Oeste": ['MS', 'MT', 'GO', 'DF'], "Norte": ['AM', 'AC', 'PA', 'RR', 'RO', 'AP', 'TO'] }
```

```python
# função para plotar cada um dos gráficos de casos/100k habitantes
def cobertura_vacinal_regiao(regiao):
  aberto_cobertura_vacinal_uf = aberto_cobertura_vacinal.groupby(["Ano", "UF","regiao"]).sum("Valor")
  aberto_cobertura_vacinal_uf = aberto_cobertura_vacinal_uf.reset_index()
  sns.set_theme(style = "whitegrid", font='serif', color_codes=True, font_scale = 1.6)

  # plotar um gráfico de regressão polinomial de ordem 2
  ax = sns.lmplot(data = aberto_cobertura_vacinal_uf[aberto_cobertura_vacinal_uf["regiao"]==regiao], x="Ano", y = "Valor", hue="UF", legend=False, height=9,  order=2, scatter_kws={"s": 50}, line_kws={"lw":5})
  plt.grid(linestyle = '--')

  # alterar a frequência dos valores do eixo y
  #plt.yticks(np.arange(25, 180, step=25))
  plt.ylim(40,160)
  plt.xticks(rotation=0)

  plt.xlabel("")
  plt.ylabel("")
  plt.title("Cobertura vacinal no {} (2001 - 2018)".format(regiao))

  ax.add_legend(label_order = ufs[regiao], fontsize=17, title="", loc='upper center', ncol=4, borderaxespad=2, frameon = True)

  # adição de linha vermelha na meta da BCG (90%)
  plt.axhline(90, color='darkviolet', linewidth=1.5, ls ='--')

  # adição do texto "meta"
  plt.text(2001.5, 86, "Meta da população-alvo da BCG (90%)", horizontalalignment='left',color=('darkviolet'),fontsize=16)

  image_name = f'cobertura_vacinal_regiao_{regiao}.png'
  graficos.append(image_name)
  plt.savefig(image_name, bbox_inches='tight',pad_inches=0.2)

  plt.show(ax)
  print("\n \n")
```

```python
from PIL import Image, ImageOps
def concat_images(image_paths, size, shape=None):
    # Open images and resize them
    width, height = size
    images = map(Image.open, image_paths)
    images = [ImageOps.fit(image, size, Image.ANTIALIAS) 
              for image in images]
    
    # Create canvas for the final image with total size
    shape = shape if shape else (1, len(images))
    image_size = (width * shape[1], height * shape[0])
    image = Image.new('RGB', image_size)
    
    # Paste images into final image
    for row in range(shape[0]):
        for col in range(shape[1]):
            offset = width * col, height * row
            idx = row * shape[1] + col
            image.paste(images[idx], offset)
    
    return image
```
